```python
import csv
import networkx as nx
from operator  import itemgetter
from networkx.algorithms import community
import matplotlib.pyplot as plt
import pandas as pd
from networkx.algorithms import bipartite
```

```python
#unweighted graph
df = pd.read_csv('PathLinker_2018_human-ppi-weighted-cap0_75.csv')
G = nx.Graph()
for i in range(10000):
        tail=df['#tail'][i] 
        head=df['head'][i]
    

        
        G.add_edges_from([(tail,head)])
```

```python
plt.figure(figsize=(20,20))
nx.draw_networkx(G, with_labels=False, node_size=2)
plt.savefig('./Graphs/un_w_graph.jpg')
plt.show()
```

```python
#Creating a agjacency matrix
A = nx.to_numpy_matrix(G)
print(A)
```

```python
#Creating a graph for adjacency matrix.
H = nx.from_numpy_matrix(A)
plt.figure(figsize=(20,20))
nx.draw(H,with_labels=False,node_color='r', node_size=2)
plt.savefig('./Graphs/Matrix.jpg')
plt.show()
```
